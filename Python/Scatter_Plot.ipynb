## 산점도(Scatter_Plot)

```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
# Load Data as Pandas.DataFrame
df = pd.read_csv('./data/iris_clabel.csv')
```

```python
# Plot sepal-length, sepal_width
df.plot.scatter(x='A', y='B')
plt.show()
```

### specify color and marker by class

```python
colors = ['r', 'g', 'b']
markers = ['.', '*', '^']
fig, ax = plt.subplots(1, 1)

for i, species in enumerate(df['Label'].unique()):
    ddf = df[df['Label']==species]
    ddf.plot.scatter(x='A', y='B',
                     alpha=0.5, s=10*(i+1), ax=ax,
                     color=colors[i], marker=markers[i], label=species)
plt.legend()
plt.show()
```

### log
- 값의 차이가 큰 경우 로그를 활용하여 관찰이 용이한 형태로 표현할 수 있다.

```python
bs = load_boston()
df = pd.DataFrame(bs.data, columns=bs.feature_names)
df['PRICE'] = bs.target  # add price
print(df.keys())

# Normal Scatter Plot
df.plot.scatter(x='CRIM', y='PRICE')
plt.title('Scatter plot of Price vs Crime')
plt.show()

# Log-X
df.plot.scatter(x='CRIM', y='PRICE', logx=True)
plt.title('Scatter plot of Price vvs log(Crime)')
plt.show()
```

## Scatter Plot Matrix

```python
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv('./data/iris.csv')
```

```python
# Scatter Plot Maxtrix with Histogram
scatter_matrix(df, alpha=0.5)
plt.show()
```

```python
# Scatter Plot Maxtrix with Kernel Density Estimation
scatter_matrix(df, alpha=0.5, diagonal='kde')
plt.show()
```
