## 바이올린 플롯(Violin Plot)

```python
# Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Iris
df = pd.read_csv('./data/iris_clabel.csv')
sns.violinplot(data=df, x='Label', y='A')
plt.title('Violin Plot of Iris Dataset')
plt.show()
```

### Tips dataset

```python
# hue를 사용해 2개의 그룹을 1개의 Violin Plot으로 표현할 수 있습니다.
tips = sns.load_dataset('tips')
sns.violinplot(x='day', y='tip', hue='sex', data=tips, split=True)
plt.show()

sns.violinplot(x='day', y='tip', hue='sex', data=tips, split=False)
plt.show()
```
