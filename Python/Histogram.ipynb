## 히스토그램(Histogram)

```python
# Import Library
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# iris data to DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Histogram
df.plot.hist(alpha=0.5)  # alpha 값으로 투명도 조절할 수 있습니다.
plt.show()
```

```python
df.plot.hist(subplots=True, layout=(2, 2), alpha=0.5, figsize=(8, 4))
plt.show()
```

```python
# bins 값으로 그래프 구간 갯수를 설정할 수 있습니다.
df.plot.hist(subplots=True, layout=(2, 2), alpha=0.5, bins=50, figsize=(8, 4))
plt.show()
```
