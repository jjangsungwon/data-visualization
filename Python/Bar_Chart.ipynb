## 막대 그래프(Bar Chart)

```python
# Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 세로(수직) 막대 그래프

```python
# Series
series = pd.Series(np.random.rand(4), index=['A', 'B', 'C', 'D'], name='Market Share')
series.plot.bar()
plt.show()
```

```python
# Data Frame
df = pd.DataFrame(np.random.rand(5, 4), columns=['A', 'B', 'C', 'D'])
df.plot.bar()
plt.show()
```

### 가로(수평) 막대 그래프

```python
# Series
series.plot.barh()
plt.show()
```

```python
# Data Frame
df.plot.barh()
plt.show()
```

### 누적 막대 그래프
- 누적 막대 그래프에서 가로, 세로는 위에 구현한 방법을 응용하면 됩니다.

```python
# Series
plt.subplot(1, 2, 1)
series.plot.bar(stacked=True)  # Vertical

plt.subplot(1, 2, 2)
series.plot.barh(stacked=True)  # Horizontal
```

```python
# Data Frame
df.plot.bar(stacked=True)  # Vertical

df.plot.barh(stacked=True)  # Horizontal
```
